```python
# importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
# checking the number of rows and columns 
data.shape
```

```python
# getting the dataset stats
data.describe()
```

```python
# checking if there are null values 
data.isnull().sum()
```

```python
# checking the number of each value in the target column
data['Class'].value_counts()
```

```python
# atributing variables for each item of target colums
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
# getting stats abour the ligit values
legit.Amount.describe()
```

```python
# getting stats abour the fraud values

fraud.Amount.describe()
```

```python
# grouping the class items to understand the stats and differences between the target values
data.groupby('Class').mean()
```

```python
# chosing among legit sample the same number of rows as fraud sample
legit_sample = legit.sample(n=492)
```

```python
# concatenate the legit sample and the fraud sameple
data_new = pd.concat([legit_sample, fraud], axis=0)
data_new.head()
```

```python
# checking if now we have the same number of fraud and legit items
data_new['Class'].value_counts()
```

```python
# grouping the class items to understand the stats and differences between the target values

data_new.groupby('Class').mean()
```

```python
# splitting the dataset in feature and target
X = data_new.drop(['Class'], axis=1)
y = data_new['Class']
```

```python
# splitting the dataset in train and test samples
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2, test_size=0.2, stratify=y)
```

```python
# importing the model
from sklearn.linear_model import LogisticRegression
```

```python
# training and predicting the train sample
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
pred = model.predict(X_train)
```

```python
# importing the metrics
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
```

```python
classification = classification_report(y_train, pred)
print(classification)
```

```python
#  predicting the test sample
pred_test = model.predict(X_test)
classification = classification_report(y_test, pred_test)
print(classification)
```

```python
roc = roc_auc_score(y_test, pred_test)
print(f'The ROC-AUC score is {roc:.2f}')
```

```python
tpr, tnr, limit = roc_curve(y_test, pred_test)
```

```python
#plotting ROC Curve graph
plt.subplots(1, figsize = (5,5))
plt.title('ROC Curve')
plt.plot(tpr, tnr)
plt.xlabel('specifity')
plt.ylabel('sensibility')
plt.plot([0,1], ls = "--", c='red') #red dotted line
plt.plot([0,0], [1,0], ls= '--', c='green'), plt.plot([1,1], ls= '--', c='green') #green dotted line
plt.show()
```
